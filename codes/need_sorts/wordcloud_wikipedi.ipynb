```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import wikipediaapi

wiki = wikipediaapi.Wikipedia('ko')
page = wiki.page('인공지능')
text = page.text

f = open("stopwords_korean.txt", "rt", encoding="utf-8")
lines = f.readlines()
stop_words = []
for line in lines:
    line = line.replace('\n', '')
    stop_words.append(line)
f.close()

wordcloud = WordCloud(colormap='winter', stopwords=stop_words).generate(text)
# wordcloud = WordCloud( background_color='white', colormap='winter', stopwords=stop_words).generate(text)

plt.figure(figsize=(10, 10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
